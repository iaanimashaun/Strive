### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
df = pd.DataFrame(data)
df.plot(x='month_number', y="total_profit")
plt.xlabel("Month Name")
plt.ylabel("Total Profit")
plt.title("Company Profit by Month")
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
num = df.drop(['month_number','total_profit', 'total_units'], axis=1).sum(axis=0)
num
```

```python
denominator = df.drop(['month_number','total_profit', 'total_units'], axis=1).sum(axis=0).sum()
denominator
```

```python
values = num/denominator*100
values
```

```python
autopct=lambda p: '{:.2f}%({:.0f})'
```

```python
plt.pie(values, autopct='%.2f%%', radius=2, explode=[0,0,0,0.1,0,0], labels=df.columns[1:7]);
plt.legend(loc='lower right', bbox_to_anchor=(2.3, -0.3), fontsize=12);
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
df.head(3)
```

```python
# Write your code here

plt.style.use('fivethirtyeight')

fig, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(12,8))

x = df['month_number']
y1 = df['bathingsoap']
y2 = df['facewash']
ax1.plot(x, y1, 'ko-')
ax1.set_title('Sales Data of Bathing Soap')
ax1.set_ylim(6000, 15000)
ax2.plot(x, y2, 'r-o')
ax2.set_title('Sales Data of Facewash')
ax2.set_ylim(1000, 2200)
ax2.set_xlabel('Month')
ax2.set_ylabel('Sales Units in Numbers')
ax2.set_xticks(range(1,13))

plt.tight_layout()
plt.show()
```
